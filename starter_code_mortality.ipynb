```python
import pandas as pd
```

```python
# Read in data for tuberculosis
tuberculosis_data = pd.read_csv(r"Data/WHOMortalityDatabase_Deaths_Tuberculosis.csv", skiprows=[0,1,2,3,4,5], index_col = False)
tuberculosis_df_orig = pd.DataFrame(tuberculosis_data)
tuberculosis_df_orig.head()
```

```python
malaria_data = pd.read_csv(r"Data/WHOMortalityDatabase_Deaths_Malaria.csv", skiprows=[0,1,2,3,4,5], index_col=False)
malaria_df_orig = pd.DataFrame(malaria_data)
malaria_df_orig.head()
```

```python
stds_data = pd.read_csv(r"Data/WHOMortalityDatabase_Deaths_STDs.csv", skiprows=[0,1,2,3,4,5], index_col=False)
std_df_orig = pd.DataFrame(stds_data)
std_df_orig.head()
```

```python
hiv_aids_data = pd.read_csv(r"Data/WHOMortalityDatabase_Deaths_HIV_AIDS.csv", skiprows=[0,1,2,3,4,5], index_col=False)
hiv_aids_df_orig = pd.DataFrame(hiv_aids_data)
hiv_aids_df_orig.head()
```

```python
tuberculosis_df = tuberculosis_df_orig
malaria_df = malaria_df_orig
std_df = std_df_orig
hiv_aids_df = hiv_aids_df_orig
```

```python
tuberculosis_df['Cause of Death'] = "Tuberculosis"
malaria_df['Cause of Death'] = "Malaria"
std_df['Cause of Death'] = "STDs (exc. HIV/AIDS)"
hiv_aids_df['Cause of Death'] = "HIV/AIDS"
```

```python
infectious_disease_df = pd.concat([tuberculosis_df, malaria_df, std_df, hiv_aids_df], axis = 0)
infectious_disease_df.head()
```

```python
infectious_disease_df.info()
```

```python
agestandardized_df = infectious_disease_df.dropna()
agestandardized_df= agestandardized_df.drop(agestandardized_df[agestandardized_df['Year'] < 2010].index)
agestandardized_df.head()
```

```python
infectious_disease_df = infectious_disease_df.drop(['Age-standardized death rate per 100 000 standard population'], axis = 1)
```

```python
infectious_disease_df['Age Group'] = infectious_disease_df['Age Group'].str.strip('[]').astype(str)
```

```python
infectious_disease_df = infectious_disease_df.drop(infectious_disease_df[infectious_disease_df['Year'] < 2010].index)
infectious_disease_df.head()
```

```python
infectious_disease_df.to_csv('Data/infectious_disease.csv')
agestandardized_df.to_csv('Data/agestandardized.csv')
```
